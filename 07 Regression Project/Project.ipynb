## Step 4 - Feature Selection

categorical - OrdinalEncoding

continous - StandardScaler

```python
cat = list(X.columns[X.dtypes == "object"])
con = list(X.columns[X.dtypes != "object"])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe1 = make_pipeline(SimpleImputer(strategy="median"), StandardScaler())
```

```python
cat_pipe1 = make_pipeline(
    SimpleImputer(strategy="constant", fill_value="Not_Avail"), OrdinalEncoder()
)
```

```python
pre1 = ColumnTransformer([("num", num_pipe1, con), ("cat", cat_pipe1, cat)]).set_output(
    transform="pandas"
)
```

```python
X_pre = pre1.fit_transform(X)
X_pre.head()
```

### Feature Selection

```python
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LinearRegression

base_model = LinearRegression()

sel = SequentialFeatureSelector(
    base_model, n_features_to_select="auto", direction="backward"
)
sel.fit(X_pre, Y)

imp_cols = sel.get_feature_names_out()
print(imp_cols)
```
