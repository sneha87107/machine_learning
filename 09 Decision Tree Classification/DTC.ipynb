## Step 3 - Seperate X and Y

```python
X = df.drop(columns=["Drug"])
Y = df[["Drug"]]
```

```python
Y.head()
```

## Step 4 - Apply Preprocessing on X

```python
X.dtypes
```

```python
cat = list(X.columns[X.dtypes == "object"])
con = list(X.columns[X.dtypes != "object"])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe = make_pipeline(SimpleImputer(strategy="median"), StandardScaler())
```

```python
cat_pipe = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore", sparse_output=False, drop="first"),
)
```

```python
pre = ColumnTransformer([("num", num_pipe, con), ("cat", cat_pipe, cat)]).set_output(
    transform="pandas"
)
```

```python
X_pre = pre.fit_transform(X)
X_pre.head()
```

```python
pre
```

## Step 5 - Train Test Split


test - 0.2 - 0.4

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(
    X_pre, Y, test_size=0.4, random_state=42
)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

## Step 6 - Model Building

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(
    max_depth=1, min_samples_split=5, min_samples_leaf=5, criterion="gini"
)

model.fit(xtrain, ytrain)
```

```python
model.score(xtrain, ytrain)
```

```python
model.score(xtest, ytest)
```

## Tuning the hyperparameters in model

```python
params = {
    "max_depth": [2, 3, 4, 5],
    "min_samples_split": [5, 6, 7, 8],
    "min_samples_leaf": [5, 6, 7, 8],
    "criterion": ["gini", "entropy"],
}
```

```python
params
```

## Hyperparameter tuning techniques

1. GridSearchCV - for tuning single value - Takes lot of time to tune because all possible combinations are checked
2. RandomisedSearchCV - for tuning multiple parameters, Paramaters are selected at random , takes less time to tune

```python
from sklearn.model_selection import RandomizedSearchCV

base_dtc = DecisionTreeClassifier()
rscv = RandomizedSearchCV(
    estimator=base_dtc, param_distributions=params, cv=5, scoring="f1_macro"
)

rscv.fit(xtrain, ytrain)
```

```python
rscv.best_params_
```

#### Everyone might get different parameters than me

```python
rscv.best_score_
```

```python
best_dtc = rscv.best_estimator_
best_dtc
```

## Step 7 - Model Evaluation

```python
# Train Accuracy
best_dtc.score(xtrain, ytrain)
```

```python
# Test Accuracy
best_dtc.score(xtest, ytest)
```

```python
ypred_test = best_dtc.predict(xtest)
ypred_test[0:5]
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_estimator(best_dtc, xtest, ytest)
```
